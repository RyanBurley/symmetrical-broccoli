```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
coordinates = pd.read_csv('catalog.csv')
```

```python
coordinates
```

```python
plt.plot(coordinates['ra'],coordinates['dec'],'o',markersize=4,color='green')
plt.xlabel('Right Ascension (deg)')
plt.ylabel('Declination (deg)')
```
