```python
def crop_to_circle(ras,decs,r_asc,dec,radius):
    '''
    Placeholder function
    '''
    cut_ras = []
    cut_decs = []

    for i in range(len(ras)):
        if ((ras[i] - r_asc)**2 + (decs[i] - dec)**2 < radius**2):
            cut_ras.append(ras[i])
            cut_decs.append(decs[i])

    return cut_ras, cut_decs
```
